```python
import torch
```

```python
import itk
import tqdm
ds = []
for tt in tqdm.tqdm(test_pair_paths):
    tt = tt.replace("name", "name-raid")
    tt = tt.split()
    data = []
    for t in tt:
        image = torch.tensor(itk.GetArrayFromImage(itk.imread(t)))
        if "RIGHT" in t:
            image = torch.flip(image, [1])
        elif "LEFT" in t:
            pass
        else:
            raise FuckityError()
        data.append(image)
    ds.append(data)
        
```

```python
ds_small = []
for iA, iB, cA, cB in tqdm.tqdm(ds):
    iA = iA[None, None, :, :, :]
    iB = iB[None, None, :, :, :]
    
    iA = torch.nn.functional.avg_pool3d(iA, 4)
    iB = torch.nn.functional.avg_pool3d(iB, 4)
    
    cA = cA[None, None, :, :, :]
    cB = cB[None, None, :, :, :]
    cA = cA.byte()
    cB = cB.byte()
    
    ds_small.append([iA, iB, cA, cB])
    
```

```python
torch.save(ds_small, "/name/name/knees_test_set")
```
