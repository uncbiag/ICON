```python
import os
ars = []
folder = "../synthetic_example_out_kernel_weighting_type_sqrt_w_K_sqrt_w/brain_affine_icbm/"
def tonumber(s):
    s = "".join([c for c in s if c in "1234567890"])
    i = int(s)
    return i
filenames = sorted(os.listdir(folder), key=tonumber)
for fname in tqdm.tqdm(filenames):
    ars.append(itk.array_from_image(itk.imread(folder + fname)))
```

```python
import numpy as np
ars2 = np.array(ars)
```
