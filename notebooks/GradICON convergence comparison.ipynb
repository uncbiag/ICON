```python
import torch
import pickle
import matplotlib.pyplot as plt
```

```python
ICON_fullres_loss_curve = pickle.load(
    open(
    "../../masked_similarity_InvCon/results/fourth_net_frozen/loss_curve"
    , "rb"))
```

```python
ICON_halfres_loss_curve = pickle.load(
    open(
    "../../masked_similarity_InvCon/results/hires_smart_6/loss_curve",
    "rb"))
```

```python
ICON_quarterres_loss_curve = pickle.load(
    open(
        "../../masked_similarity_InvCon/results/smart_lmbda_4/loss_curve", 
        "rb"))
```

```python
import matplotlib.pyplot as plt
```

```python
print("Similarity")
plt.plot([x[1] for x in ICON_quarterres_loss_curve])
```

```python
print("Transform Magnitude")
plt.plot([x[2] for x in ICON_quarterres_loss_curve])
```

```python
print("Folds")
plt.plot([x[3].mean().item() for x in ICON_quarterres_loss_curve])
```

```python
GradICON_quarterres_loss_curve = pickle.load(
    open("../results/gradknee-2/loss_curve", "rb"))
GradICON_halfres_loss_curve = pickle.load(
    open("../results/grad_hires-1/loss_curve", "rb"))
GradICON_fullres_loss_curve = pickle.load(
    open("../results/grad_fullres-6/loss_curve", "rb"))
```

```python
print("Transform_magnitude")
plt.plot([x[2] for x in ICON_quarterres_loss_curve])
plt.plot([x[2] for x in GradICON_quarterres_loss_curve])
```

```python
print("Similarity")
plt.plot([x[1] for x in ICON_quarterres_loss_curve])
plt.plot([x[1] for x in GradICON_quarterres_loss_curve])
```

```python
print("Folds")
plt.plot([x[3].mean().item() for x in ICON_quarterres_loss_curve])
plt.plot([x[3].mean().item() for x in GradICON_quarterres_loss_curve])
```

```python
ICON_whole = ICON_quarterres_loss_curve + ICON_halfres_loss_curve + ICON_fullres_loss_curve
GradICON_whole = GradICON_quarterres_loss_curve + GradICON_halfres_loss_curve #+ GradICON_fullres_loss_curve

print("squared Transform_magnitude")
plt.plot([x[2] for x in ICON_whole], )
plt.plot([x[2] for x in GradICON_whole])
```

```python
print("Similarity_measure")
plt.plot([x[1] for x in ICON_whole])
plt.plot([x[1] for x in GradICON_whole])
```

```python
print("Folds")
plt.plot([x[3].mean().item() for x in ICON_whole])
plt.plot([x[3].mean().item() for x in GradICON_whole])
```

```python
plt.subplot(2, 2, 1)
plt.title("similarity loss")
plt.plot([x[1] for x in ICON_whole])
plt.plot([x[1] for x in GradICON_whole])
plt.subplot(2, 2, 2)
plt.title("transform magnitude")
plt.plot([x[2] for x in ICON_whole], )
plt.plot([x[2] for x in GradICON_whole])
plt.xlabel("step")
plt.subplot(2, 2, 3)
plt.title("log folds")
plt.plot([torch.log(x[3].mean()).item() for x in ICON_whole])
plt.plot([torch.log(x[3].mean()).item() for x in GradICON_whole])
plt.tight_layout()
plt.subplot(2, 2, 4)
plt.plot([], label="ICON")
plt.plot([], label="GradICON")
plt.xticks([])
plt.yticks([])
plt.legend()
plt.show()
```
